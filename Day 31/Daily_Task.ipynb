```python
#For loop with list
list=['C','C++','Java','C#']
print("This is our list ",list)
i=1
for val in list:
 print("List Item",i,"is",val)
 i=i+1 
```

```python
#For Loop with Tuple
tuple=('C','C++','Java','C#','Python')
print("This is our Tuple ",tuple)
i=1 
for val in tuple:
 print("Tuple Item",i,"is",val)
 i=i+1
```

```python
#For loop with else condition
#program to check weather the given number is prime number or not
num=int(input("Enter a Number : "))
for i in range(2,num):
    if(num%i==0):
        print("%d is not a prime number..."%num)
        break
        print("%d is a prime number..."%num)
                        
```

```python
str=""
for Row in range(0,7):
    for Col in range(0,7):
        if(((Col == 1 or Col == 5) and Row != 0) or ((Row == 0 or Row == 3) and (Col > 1 and Col < 5))):
            str=str+"*"
        else:
                str=str+""
                str=str+"\n"
                print(str)
```
